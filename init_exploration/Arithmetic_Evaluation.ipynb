```python
import openai
from dotenv import find_dotenv, load_dotenv
import pickle

load_dotenv(find_dotenv())
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate

def parse_query_with_llm(question, context):
    
    template = """You are an assistant that parses mathematical Questions.
    
    Given the context and Question, extract:
    - The mathematical operation (e.g., sum, average, difference, etc.)
    - The values involved
    - The formula to compute the result
    
    Current context:
    {context}
    
    Current Question: {question}
    
    Please provide a structured response with:
    Operation: <operation>
    Values: <list of values>
    Formula: <formula>
    Answer: <answer>
    
    KEEP RESPONSES CONCISE AND FOCUSED.
    ONLY USE INFORMATION PROVIDED IN THE CONTEXT.
    """
    
    # Create prompt template
    prompt = ChatPromptTemplate.from_template(template)
    
    # Initialize LLM
    model = ChatOpenAI(
        temperature=0,
        model="gpt-4o-mini",
        max_tokens=200  # Limit response length
    )
    
    # Format prompt with input variables
    formatted_prompt = prompt.format_messages(
        context=context,
        Question=question
    )
    
    # Get response from LLM
    response = model(formatted_prompt)
    
    # Parse the response into structured format
    response_lines = response.content.strip().split('\n')
    parsed_response = {}
    
    for line in response_lines:
        if ':' in line:
            key, value = line.split(':', 1)
            parsed_response[key.strip()] = value.strip()
    
    return parsed_response
```

```python
# Parse the query
parsed_response = parse_query_with_llm(df["qa"].get("question"), df["table_ori"])
print("Parsed Response:", parsed_response)
print("Answer:", (parsed_response["Answer"]))
```

```python
def evaluate_model(model_function, test_cases):
    results = []
    
    for _, test in test_cases.iterrows():  # Use iterrows() to get row data
        # Get model prediction
        predicted = model_function(test["qa"]["question"], test["table_ori"])
        
        # Compare with expected output
        correct = (predicted["Answer"]) == test["qa"].get("answer")
        print("{0}, {1}".format((predicted["Answer"]), test["qa"].get("answer")))
        results.append(correct)
    
    # Compute accuracy
    accuracy = sum(results) / len(results)
    return accuracy, results

# Run evaluation
accuracy, results = evaluate_model(parse_query_with_llm, df)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
print()
```
